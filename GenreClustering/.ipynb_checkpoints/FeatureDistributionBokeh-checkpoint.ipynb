```python
from bokeh.models.callbacks import CustomJS

track_genres = pd.read_csv('clusters.csv', index_col=0)

audio_cols = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'speechiness', 'tempo', 'valence']

data=df
# Create data source
source = ColumnDataSource(data={
    'x' : data['acousticness'],
    'y' : data['energy'],
    'track' : data['track'],
    'artist' : data['artist'],
    'album' : data['album'],
    'genres_top' : data['genres_top'],
    'genres_all' : data['genres_all']
})

source_clusters = ColumnDataSource(data=track_genres)

unique_genres = data['genres_top'].unique().tolist()
color_mapper = CategoricalColorMapper(factors=unique_genres, palette=Category20_17)

# Add hover tool
# hover = HoverTool(tooltips={'Artist':'@artist',
#                                      'Track':'@track',
#                                      'Album':'@album',
#                                      'Top Genre': '@genres_top',
#                                      'All Genres':'@genres_all'})
# Create figure
p = figure(
           x_range = (0,1.0), y_range = (0,1.0),
           plot_height=400, plot_width=600, 
           )

# Plot Points
tracks_plot = p.circle(x='x', y='y', source=source, fill_alpha=0.3, legend='genres_top', 
         color = dict(field='genres_top', transform=color_mapper))

# # Create a dropdown Select widget for the x data: x_select
x_select = Select(
    options=audio_cols,
    value='acousticness',
    title='x-axis data'
)

# # Attach the update_plot callback to the 'value' property of x_select
x_select.on_change('value', update_plot)

# # Create a dropdown Select widget for the y data: y_select
y_select = Select(
    options=audio_cols,
    value='energy',
    title='y-axis data'
)

# Attach the update_plot callback to the 'value' property of y_select
y_select.on_change('value', update_plot)

# Plot Clusters
p.circle(x=x_select.value, y=y_select.value, source=source_clusters, fill_alpha=0.2, size=100,
        color=dict(field='cluster', transform=color_mapper))
### Trick to locate legend outside of plot
new_legend = p.legend[0]
p.legend[0].plot = None
p.add_layout(new_legend, 'right')




callback = CustomJS(args=dict(source=source), code="""
    var data = source.data;
    x = data['x']
    y = data['y']
    source.change.emit();
""")

x_select.js_on_change('value', callback)
y_select.js_on_change('value', callback)

# # Create layout and add to current document

layout = column(p, widgetbox(x_select, y_select))
curdoc().add_root(layout)
show(layout)
```
