# Tracks By Genre And Year

This notebook will explore some interesting trends and insights from the [fma dataset](https://github.com/mdeff/fma/). We will be using the files 'genres.csv' and 'tracks.csv' in our analysis. These files are described as the following:
- tracks.csv: per track metadata such as ID, title, artist, genres, tags and play counts, for all 106,574 tracks.
- genres.csv: all 163 genre IDs with their name and parent (used to infer the genre hierarchy and top-level genres).

First, some necessary imports

```python
%matplotlib inline

import ast
import os
import IPython.display as ipd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Built-in module from FMA
import utils
```

```python
### Creating genres dataframe from csv using utils from fma repo

genres = utils.load('C:\\Users\\ramir\\OneDrive\\Documents\\GitHub\\fma_metadata\\genres.csv')
```

```python
genres.info()
```

Extracting data from the tracks spreadsheet is more elaborate. 

```python
filepath='C:\\Users\\ramir\\OneDrive\\Documents\\GitHub\\fma_metadata\\tracks.csv'
tracks = pd.read_csv(filepath, index_col=0, header=[0, 1])


SUBSETS = ('small', 'medium', 'large')
tracks['set', 'subset'] = tracks['set', 'subset'].astype(
        'category', categories=SUBSETS, ordered=True)
    
### Basic columns for data analysis
basic_columns = [('album','date_released'), ('album', 'title'), ('album', 'id'),
                ('artist', 'id'), ('artist', 'name'),  
                 ('track', 'genres'), ('track', 'genres_top'), ('track', 'genres_all'),
                 ('track', 'title'), ('track', 'listens')]

### Columns that should be in datetime format
COLUMNS_dt = [('album', 'date_released')]

### Set tracks df to just basic columns
tracks = tracks[basic_columns]

### Remove rows with blank genres
tracks = tracks[~tracks[('track','genres_all')].isnull()]

### Remove rows with genres that don't have genre codes, which are mislabeled. 
tracks.drop(tracks[tracks[('track', 'genres_all')].str.contains('[a-z]')].index, inplace=True)

### Replace blank top genres with 'NONE'
tracks[('track', 'genres_top')].fillna('UNKNOWN', inplace=True)

### Assign datetime columns to datetime type
for column in COLUMNS_dt:    
    times = pd.to_datetime(tracks[column], errors='coerce')
    years = [time.year for time in times]
    tracks.loc[:,column] = years
    
    # Assume that any artist that doesn't have an active year end is still active
    if (column == ('artist', 'active_year_end')):
        
        tracks[column].fillna(pd.to_datetime('today').year, inplace=True)
        
    else:
        tracks[column].fillna(0, inplace=True)
        
### Columns that should be categories
COLUMNS_cat = [('track', 'genres_top')]

for column in COLUMNS_cat:
    tracks[column] = tracks[column].astype('category')
    
    
### Columns that should be numeric
COLUMNS_int = [('track', 'listens')]
for column in COLUMNS_int:
    tracks[column] = pd.to_numeric(tracks[column], errors='coerce')
    tracks[column].fillna(0, inplace=True)

    
tracks.head(10)
```

```python
tracks.info()
```

#### Categorizing Tracks By Release Year

```python
def categorize_year(year):
    if 2010 <= year <= 2018:
        year = '2010\'s'
    elif 2000 <= year <= 2009:
        year = '2000\'s'
    elif 1990 <= year <= 1999:
        year = '1990\'s'
    elif 1980 <= year <= 1989:
        year = '1980\'s'
    elif 1970 <= year <= 1979:
        year = '1970\'s'
    elif 1960 <= year <= 1969:
        year = '1960\'s'
    else:
        year = 'UNKNOWN'
    return year

tracks[('track', 'release_decade')] = [categorize_year(y) for y in tracks[('album', 'date_released')]]
tracks[('track', 'release_decade')].astype('category')
tracks.info()
```

### Counting Tracks and Genres

Since the this project is about genre classification and clustering, it's important to examine how the genres and subgenres in the dataset are distributed. 

First, we're going to count how many tracks there are per genre. We'll be analyzing the genres dataframe. Note that many tracks fall under multiple genres, so summing up all the genres won't equal to the number of tracks total.

```python
genres.sort_values('#tracks', ascending=False, inplace=True)
genres.head(10)
```

Plotting every genre will yield an ugly plot considering there are 163 genres.

```python
genres.plot(x='title', y='#tracks', kind='bar')
```

Let's revise the plot so we see only the 'main' genres. This represents the genres where 'parent' = 0, meaning that the genre is at the top of its hierarchy. Alternatively, this can also be done by using the conditional 'top_level' = 'genre_id'. 

```python
genres_main = genres[genres['top_level']==genres.index]
genres_main.plot(x='title', y='#tracks', kind='bar', sort_columns=True)
```

Then we can explore the subgenres within a single genre. Let's use Rock as an example.

```python
genres_rock = genres[genres['parent']==12]
genres_rock.plot(x='title', y='#tracks', kind='bar', sort_columns=True, title='Rock Subgenres')
```

The most popular subgenre of rock is Punk while Rock Operas are the least common subgenre.

We can do this for other genres as well. 

```python
genres_exp = genres[genres['parent']==38]
genres_exp.plot(x='title', y='#tracks', kind='bar', sort_columns=True, title='Experimental Subgenres')
```

```python
genres_el = genres[genres['parent']==15]
genres_el.plot(x='title', y='#tracks', kind='bar', sort_columns=True, title='Electronic Subgenres')
```

### Number of tracks over time

Next, we'll use the tracks.csv and examine tracks by release year. A track's release year is a significant distinguishing factor stylistically. 70's rock sounds different 2010's modern rock for example. Cultural trends and available technology in recording equipment are among many factors that account for this. 

```python
tracks_by_release = tracks[('album', 'date_released')].groupby(by=tracks[('album', 'date_released')].values).agg('count')
tracks_by_release.drop(0.0, inplace=True)
y = tracks_by_release.values
x = tracks_by_release.index
plt.scatter(x,y)
plt.xlim(1960,2018)
plt.xlabel('Year Released')
plt.ylabel('Number of Tracks')
plt.title('Number of Tracks By Release Year')
```
